```python
cat_columns = df.select_dtypes('object').columns
for col in cat_columns:
  print(f"{col} unique values:\n", df[col].value_counts(), "\n")
```

```python
import numpy as np

# Select only numeric columns
df_numeric = df.select_dtypes(include=['number'])

# Compute Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)

# Compute IQR
IQR = Q3 - Q1

# Define outliers (1.5 times IQR rule)
outliers = ((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR)))

# Count the number of outliers in each column
outlier_counts = outliers.sum()

# Show columns with the most outliers
outlier_counts[outlier_counts > 0].sort_values(ascending=False)
```
