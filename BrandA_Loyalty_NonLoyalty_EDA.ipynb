#** Basic Data Cleaning Implementation **

Basic Data cleaning-


1.Removing null transaction ids.


2.Change datatype of itemid field from number to string.


3.Some values are missing in customer_nationality column, dealing with it.


4.Converting item id to string.

```python
#Remove rows with null transaction IDs
df_cleaned = df.dropna(subset=['transactionId'])


#changing the datatype of itemidto string(categoriccal data)
df_cleaned['itemid'] = df_cleaned['itemid'].astype(str)

#filling missing values in customer_nationality column with unknown
df_cleaned['customer_nationality'].fillna('Unknwon', inplace = True)

#converting item_id into categorical data type()memory effeciency
df_cleaned['itemid'] =df_cleaned['itemid'].astype('category')
```

#** Feature Engineering:**

**Client has given some information regarding the fields on that basis we will perform feature engineering**




1 Missing customer_id and signed_up_loyalty_program_date


 Issue: Some transactions have customer_id but no loyalty sign-up date, while others have no customer_id at all(null values are non registered usrs given infromation from client )

 Solution: Creating a Registered field to classify customers as "Registered" or "Not Registered" for proper segmentation.



2 Negative total_spend values


 Issue: Some transactions have negative total spend, indicating refunds, while others have zero spend but positive quantity, meaning freebies.


 Solution: Creating a Refund_Transaction field to classify transactions as "Sales", "Refund", or "Freebie" to track revenue correctly.

3 Negative quantity values


 Issue: Some rows have negative quantities, suggesting refunded items rather than full transaction refunds.


 Solution: Creating an Item_Refunded field to mark refunded items separately, helping in product return analysis.

4 Missing values in currentTier


 Issue: Some customers have no tier assigned, likely indicating they arent part of the loyalty program.


 Solution: Filling missing values with "No current tier" to prevent issues in filtering and grouping.


```python
# Creating a field to check if a customer is registered

df_cleaned['Registered'] = df_cleaned.apply(lambda x: 'Registered' if pd.notnull(x['customer_id']) and pd.notnull(x['signed_up_loyalty_program_date']) else 'Guest Customer', axis = 1)
df_cleaned.head()
```

```python
df_cleaned['Registered'].value_counts()
```

```python
# Creating a field to checks if the transaction was a refund or not
def classify_transactions(spend,quantity):
  if spend < 0:
    return 'Refund'
  elif spend == 0 and quantity > 0:
    return 'Freebie'
  else:
    return 'Sales'

df_cleaned['Refund_Transactions'] = df_cleaned.apply(lambda x: classify_transactions(x['total_spend'], x['quantity']), axis = 1)
```

```python
df_cleaned['Refund_Transactions'].value_counts()
```

```python
#Creating a field to check if an item was refunded or not

df_cleaned['Item_Refunded'] = df_cleaned['quantity'].apply(lambda x: 'Refunded' if x < 0 else 'Not Refunded')
```

```python
df_cleaned['Item_Refunded'].value_counts()
```

```python
#Replacing nulls in currentTier with 'No current tier'

df_cleaned['currentTier'].fillna('No Current Tier', inplace= True)
```
